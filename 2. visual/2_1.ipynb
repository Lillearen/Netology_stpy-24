```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 13,6
```

```python
data = pd.read_csv('BankChurners.csv')
data
```

```python
data.drop(data.columns[[-1, -2, -4, -7]], axis='columns', inplace = True)
```

```python
[column for column in data.columns]
```

```python
data.describe()
```

```python
Income_Category_Credit_Limit = data[['Income_Category', 'Credit_Limit']].groupby(by='Income_Category').mean()
Income_Category_Credit_Limit
```

```python
Income_Category_Credit_Limit.loc[['Less than $40K', '$40K - $60K', '$60K - $80K', '$80K - $120K', '$120K +', 'Unknown']].plot(kind='bar', title='Средний кредитный лимит в зависимости от годового дохода клиента')
```

Наблюдаем прямую корреляцию кредитного лимита и дохода клиента. Что логично с точки зрения банка, так как кредитный лимит устанавливается на основе платежеспособности его получателя.

```python
plt.plot(data.Customer_Age, data.Months_on_book, 'o')
plt.title('Период взаимодействия с банком')
plt.xlabel('Возраст клиента')
plt.ylabel('Количество месяцев взаимодействия')
```

Чем выше возраст клиента, тем большее количество месяцев он взаимодействовал с банком - здесь тоже прямая корреляция.

```python
Age_avg_creadit_limit = data[['Customer_Age', 'Credit_Limit']].groupby('Customer_Age').mean()
Age_avg_creadit_limit.head()
```

```python
plt.plot(Age_avg_creadit_limit.index, Age_avg_creadit_limit.Credit_Limit)
plt.title('Средняя сумма кредитного лимита по возрасту клиента')
plt.xlabel('Возраст клиента')
plt.ylabel('Средняя сумма кредитного лимита')
```

Наблюдаем интересное влияние возраста клиента, на выданный ему кредитный лимит.
У молодых клиентов (в районе 25 лет) наблюдаем пик. Возможно, это связано с тем, что в этом возрасте молодые люди обычно заканчивают образование и получают более высокооплачиваемую и стабильную работу.
Можно предположить, что среди возрастных клиентов больше людей с продолжительной кредитной историей, которым банк доверяет больше, либо больше состоятельных людей.
У совсем пожилых кредит снижается, вероятно, из-за снижения ожидаемой продолжительности жизни.

```python
Number_clients_mar_status = data[['CLIENTNUM', 'Marital_Status']].groupby('Marital_Status').count()
Number_clients_mar_status
```

```python
Number_clients_mar_status.loc[['Single', 'Married', 'Divorced', 'Unknown']].plot(kind='bar', title='Количество клиентов в разном семейном положении')
```

Больше всего женатых клиентов. Либо у людей в этом статусе больше потребностей, которые проще либо необходимо удовлетворять через услуги банка, либо данный статус в принципе является наиболее распространенным среди населения.
Мало разведенных клиентов. Возможно потому, что находясь в этом статусе люди предпочитают указывать, что они одиноки.
Довольно мало клиентов с неизвестным семейным положением, что может свидетельствовать о хорошем качестве сбора данных в изучаемом банке.
