```python
import pandas as pd
import numpy as np
import re
```

```python
horses = pd.read_csv('horses.csv', header=None,  na_values='?')
horses.head()
```

```python
columns_names = dict()

with open('horses_metadata.txt') as f:
    for line in f:
        found = re.findall(r'(\d{1,2}):\s+((\w+.\s+)+)', line)
        if found:
            columns_names[int(found[0][0])-1] = found[0][1].strip()

columns_names
```

```python
selected_columns = columns_names[0], columns_names[1], columns_names[2], columns_names[3], columns_names[4], columns_names[5], columns_names[15], columns_names[22]
selected_columns
```

```python
chosen_data = horses.loc[:, [0, 1, 2, 3, 4, 5, 15, 22]]
chosen_data
```

```python
chosen_data.columns = selected_columns
chosen_data
```

```python
chosen_data.describe()
```

```python
chosen_data.max() - chosen_data.min()
```

```python
for column in ['rectal temperature', 'pulse', 'respiratory rate', 'nasogastric reflux PH']:
    print(column.upper())
    print('минимум:', chosen_data[column].min())
    print('максимум:', chosen_data[column].max())
    print('размах:', chosen_data[column].max() - chosen_data[column].min())
    print('среднее значение:', chosen_data[column].mean())
    print('медиана:', chosen_data[column].median())
    print('мода:', chosen_data[column].mode()[0])
    print('среднее отклонение:', chosen_data[column].std())
    print('\n')
```

```python
q1 = chosen_data['pulse'].quantile(0.25)
q3 = chosen_data['pulse'].quantile(0.75)
iqr = q1 - q3
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

print(lower_bound, upper_bound)
```

```python
bounds = dict()

for column in ['rectal temperature','pulse', 'respiratory rate', 'nasogastric reflux PH']:
    print(column.upper(),)
    q1 = chosen_data[column].quantile(0.25)
    q3 = chosen_data[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    bounds[column] = (lower_bound, upper_bound)
    print('нижняя граница:', lower_bound)
    print('верхняя граница:', upper_bound)
    print('среднее значение:', chosen_data[column].mean())
    print('среднее значение без выбросов:', chosen_data[(chosen_data[column] > lower_bound) & (chosen_data[column] < upper_bound)][column].mean())
    print('\n')
```

Средние значения при удалении выбросов существенно не изменяются.

```python
for column in chosen_data.columns:
    pct_missing = chosen_data[column].isnull().mean()
    print(f'{column}: {pct_missing :.2%}')
```

**surgery?** - отсутствует небольшое кол-во данных и они не связаны с остальными данными. Удаляем строки с пропусками.

**Age** - не содержит пропуски.

**Hospital Number** - не содержит пропуски.

**rectal temperature** - пропущен большой процент данных. Так как это непрерывная величина, дозаполним пропуски медианой (практически совпадает со средним, выбор между ними не очень существенен).

**pulse** - отсутствует существенный процент данных. Является непрерывной величиной с большим разбросом, пропуски дозаполняем средним значением.

**respiratory rate** - содержит много пропусков и имеет большой размах. В рамках данной задачи необходимо создать набор данных без пропусков, поэтому этот столбец лучше будет просто удалить. Дозаполнение может привести к сильному искажению данных.

**nasogastric reflux PH** - столбец практически не заполнен, поэтому его тоже удаляем.

**outcome** - отсутствует небольшое кол-во данных и они не связаны с остальными данными. Удаляем строки с пропусками.

```python
data_no_nan = chosen_data.drop(['respiratory rate', 'nasogastric reflux PH'], axis=1)
```

```python
data_no_nan.drop(data_no_nan.loc[pd.isnull(data_no_nan['surgery?']), :].index, inplace = True)
data_no_nan.drop(data_no_nan.loc[pd.isnull(data_no_nan['outcome']), :].index, inplace = True)
```

```python
data_no_nan['rectal temperature'].fillna(chosen_data['rectal temperature'].median(), inplace=True)
```

```python
data_no_nan['pulse'].fillna(chosen_data['pulse'].mean(), inplace=True)
```

```python
for column in data_no_nan.columns:
    pct_missing = data_no_nan[column].isnull().mean()
    print(f'{column}: {pct_missing :.0%}')
```
